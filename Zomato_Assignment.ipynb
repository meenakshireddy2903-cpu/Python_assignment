**2. Data Cleaning**

**Performing necessary cleaning steps:**

```python
#Removing Unamed : 0 column
df = df.drop("Unnamed: 0",axis = 1)
df
```

```python
#Drop duplicate rows
df.drop_duplicates(inplace = True)
df
```

```python
columns_to_drop = ["url","phone","dish_liked","reviews_list","menu_item"]
existing_columns = [col for col in columns_to_drop if col in df.columns]
df = df.drop(existing_columns, axis=1)
df
```

```python
df.info()
```

**Handling Address column**

```python
#Handling Address Column
df["address"].unique() #It has 0 null values
```

**Handling Name column**

```python
#Handling Name Column
df["name"].unique() #It has 0 null values
```

**Handling Online_order column**

```python
#Handling Online_order column
df["online_order"].unique() #It has 0 null values
```

**Handling book_table column**

```python
#Handling Book_table column
df["book_table"].unique() #It has 0 null values
```

**Handling Rate column**

```python
#Handling Rate column
df["rate"].unique()
# Replacing unwanted values with numpy.nan
df["rate"] = df["rate"].replace("NEW", np.nan)
df["rate"] = df["rate"].replace("-", np.nan)

# Removing '/5' from rating values.,.str.replace()
df["rate"] = df["rate"].astype(str).str.replace("/5", "")

# Converting rating column to float datatype.
df["rate"] = pd.to_numeric(df["rate"], errors='coerce')
# Checking datatype of rate column
df["rate"].dtype
# Checking number of missing values in rate column
df["rate"].isnull().sum()
#Rechecking unique values in rate column
df["rate"].unique()
# Filling missing rating values with average rating (3.7)
df["rate"] = df["rate"].fillna(3.7)
# Checking unique values again after filling null values
df["rate"].unique()
```

**Handling Votes column**

```python
#Handling Votes Column
df["votes"].unique() #It has 0 null values
```

**Handling Location column**

```python
#Handling Location column
df["location"].unique()
# Checking missing values in location column
df["location"].isnull().sum()
# Removing rows where location is missing
df.dropna(subset=["location"], inplace=True)
df["location"].isnull().sum() #Removed null values
```

**Handling rest_type column**

```python
#Dropping restype column as it is not required for analysis
df = df.drop("rest_type", axis = 1)
```

**Handling Cuisines column**

```python
#Handling Cuisines column
df["cuisines"].unique()
#Checking the no.of null values in cuisines column
df["cuisines"].isnull().sum()
#Checking most repeated cuisine
df["cuisines"].mode()
#Filling null values with most repeated value
df["cuisines"] = df["cuisines"].fillna("North Indian")
df["cuisines"].isnull().sum() #Removed null values
```

**Handling Approx_cost column**

```python
#Handling approx_cost(for two people) column
df["approx_cost(for two people)"].unique()
#Checking the no.of null values in approx_cost(for two people) column
df["approx_cost(for two people)"].isnull().sum()

# Convert the column to string type
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].astype(str)

# Removing commas from cost column
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].str.replace(",", "")

# Converting cost column to float.
df["approx_cost(for two people)"] = pd.to_numeric(df["approx_cost(for two people)"], errors='coerce')

# Dropping rows where cost is missing
df.dropna(subset=["approx_cost(for two people)"], inplace=True)
df["approx_cost(for two people)"].isnull().sum() #Removed null values
```

**Handling listed_in(type) Column**

```python
#Checking unique values
df["listed_in(type)"].unique()
#Checking null values
df["listed_in(type)"].isnull().sum() # It has 0 null values
#Renaming listed_in(type) to rest_type
df.rename(columns={"listed_in(type)": "rest_type"}, inplace=True)
```

**Handling listed_in(city) column**

```python
#Checking  unique values
df["listed_in(city)"].unique()
#Checking null values
df["listed_in(city)"].isnull().sum() #It has 0 null values
```

#TASK 2: UNIVARIATE ANALYSIS

**1. Top 5 Locations with Highest Restaurants**

```python
# Counting number of restaurants in each location
top_five_loc = df["location"].value_counts().head(5)
#Plotting bar chart
plt.bar(top_five_loc.index, top_five_loc.values)
plt.title("Top 5 Locations with Highest Number of Restaurants")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

**Interpretation:**
The above chart shows the top 5 locations with the highest number of restaurants.
The location BTM has highest no.of restaurants.

**2. Distribution of restaurant ratings**

```python
# Plotting distribution of restaurant ratings
plt.hist(df["rate"], bins=10, color="red", edgecolor="black")
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.show()
```

**Interpretation:**
Most restaurants have ratings between 3.5 and 4.5.
Very few restaurants have ratings below 3.0.
Overall majority of restaurants maintain good ratings.

**3ï¸.Percentage of Restaurants Offering Online Orders**

```python
# Counting online order availability
online = df["online_order"].value_counts()
# Plotting pie chart
plt.pie(online.values, labels=online.index, autopct="%1.1f%%")

plt.title("Percentage of Restaurants Offering Online Orders")

plt.show()
```

**Interpretation:**

The chart shows the percentage of restaurants that offer online ordering services.
Majority of restaurants provide online order facility.
This indicates that online deliveris are more.

**4.Distribution of Average Cost for Two**

```python
# 4)Distribution of average cost for two

plt.figure(figsize=(8,5))
sns.histplot(df["approx_cost(for two people)"],bins = 30, kde=True)
plt.xlabel("Cost for Two")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Average Cost for Two")
plt.show()
```

**Interpretation:**

Most restaurants are mid-range.
Only a few restaurants are very high-cost.
This suggests customers prefer mid-range pricing.

# TASK 3: BIVARIATE ANALYSIS

### **Analysing the relationships**



   **Rating vs Cost**







```python
# Scatter plot for Rating vs Cost
plt.scatter(df["approx_cost(for two people)"], df["rate"], color="red")

plt.title("Rating vs Cost")
plt.xlabel("Cost for Two")
plt.ylabel("Rating")

plt.show()
```

**Conclusion:**

The scatter plot shows the relationship between cost and rating.
Many medium-cost restaurants also have high ratings.

**Rating vs Online Order**

```python
# Boxplot for Rating vs Online Order
sns.boxplot(x="online_order", y="rate", data=df, color="red")

plt.title("Rating vs Online Order")
plt.xlabel("Online Order")
plt.ylabel("Rating")

plt.show()
```

**Conclusion:**
Online availability may positively impact customer satisfaction.

**Votes vs Rating**

```python
# Scatter plot for Votes vs Rating
plt.scatter(df["votes"], df["rate"], color="red")

plt.title("Votes vs Rating")
plt.xlabel("Votes")
plt.ylabel("Rating")

plt.show()
```

**Conclusion:**

Restaurants with higher votes generally have higher ratings.
This suggests popular restaurants often maintain good quality.
